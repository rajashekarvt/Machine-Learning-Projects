```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
from sklearn.linear_model import LinearRegression;
```

```python
df=pd.read_csv('Position_Salaries.csv')
X=df.iloc[:,1:2].values
Y=df.iloc[:,-1].values
```

```python
linreg=LinearRegression()
model=linreg.fit(X,Y)
from sklearn.preprocessing import PolynomialFeatures;
polyreg=PolynomialFeatures(degree=4)
x_poly=polyreg.fit_transform(X)

linreg2=LinearRegression()
model2=linreg2.fit(x_poly,Y)
model.score(X,Y)*100
model2.score(x_poly,Y)*100

```

```python

plt.scatter(X,Y,color='red',marker='*',label='Sample Data')
plt.plot(X,linreg.predict(X),color='blue',label='Linear Regression')
plt.xlabel('Salary')
plt.ylabel('Job Level')
plt.title('Truth Or Bluff')
plt.legend(facecolor='white',fontsize=11)
plt.show()
```

```python
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color='red',label='Sample Data',marker='+')
plt.plot(X_grid,linreg2.predict(polyreg.fit_transform(X_grid)),color='green',label='Polynomial regression')
plt.xlabel('Salary')
plt.ylabel('Job Level')
plt.title("Truth Or Bluff")
plt.legend(facecolor='white',fontsize=11)
plt.show()
```
