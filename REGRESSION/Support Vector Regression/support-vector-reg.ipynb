```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
from sklearn.svm import SVR;
from sklearn.preprocessing import StandardScaler;
```

```python
df=pd.read_csv('Position_Salaries.csv')
X=df.iloc[:,1:2].values
Y=df.iloc[:,2].values
Y=Y.reshape(-1,1)
```

```python
sc_X=StandardScaler()
sc_Y=StandardScaler()
X=sc_X.fit_transform(X)
Y=sc_Y.fit_transform(Y)
```

```python
regressor=SVR(kernel='rbf') # where rbf is a gaussian Kernel it is default btw
model=regressor.fit(X,Y)
```

```python
y_pred=sc_Y.inverse_transform([regressor.predict(sc_X.fit_transform(np.array([[6.5]])))])
y_pred
```

```python


plt.scatter(X,Y,color='red',marker='*',label='Sample Data')
plt.plot(X,regressor.predict(sc_X.fit_transform(np.array(X))),color='blue',label='Linear Regression')
plt.xlabel('Salary')
plt.ylabel('Job Level')
plt.title('Truth Or Bluff')
plt.legend(facecolor='white',fontsize=11)
plt.show()
```

```python

X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color='red',marker='*',label='Sample Data') # for smoother curve we need to use np.arange to get a higher resolution / visualization
plt.plot(X_grid,regressor.predict(X_grid),color='blue',label='Linear Regression')
plt.xlabel('Salary')
plt.ylabel('Job Level')
plt.title('Truth Or Bluff')
plt.legend(facecolor='white',fontsize=11)
plt.show()
```
