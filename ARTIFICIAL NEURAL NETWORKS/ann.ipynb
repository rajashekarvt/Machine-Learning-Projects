```python
import numpy as np;

import pandas as pd;
```

```python
data=pd.read_csv("Churn_Modelling.csv")
X=data.iloc[:,3:13].values
Y=data.iloc[:,13].values
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder;
encoder1=LabelEncoder()
encoder2=LabelEncoder()
X[:,1]=encoder1.fit_transform(X[:,1])
X[:,2]=encoder2.fit_transform(X[:,2])
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.preprocessing import StandardScaler;
sc_X=StandardScaler()
X=sc_X.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split;
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
import keras;
from keras.models import Sequential;
from keras.layers import Dense;     
```

```python
classifier=Sequential()
classifier.add(Dense(units=6,activation='relu'))#first hidden layer,#relu means rectifier
classifier.add(Dense(units=6,activation='relu'))#second hidden layer
classifier.add(Dense(units=1,activation='sigmoid'))#Output Layer
classifier.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
classifier.fit(x_train,y_train,batch_size=10,epochs=100)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred=(y_pred>0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix;
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score;
accuracy_score(y_test,y_pred)
```
